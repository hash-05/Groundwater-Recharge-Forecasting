```python
def impute_zeros_with_average(df, num_closest=3):
    def get_nearest_non_zero_indices(arr):
        non_zero_indices = np.where(arr != 0)[0]
        return np.argpartition(np.abs(non_zero_indices - np.arange(len(arr))), num_closest)[:num_closest]

    imputed_df = df.copy()

    for index, row in df.iterrows():
        non_zero_indices = np.where(row != 0)[0]

        for i, value in enumerate(row):
            if value == 0:
                closest_non_zero_indices = get_nearest_non_zero_indices(non_zero_indices - i)
                closest_non_zero_values = row[non_zero_indices[closest_non_zero_indices]]
                imputed_df.at[index, df.columns[i]] = np.mean(closest_non_zero_values)

    return imputed_df
```

```python
df_Oct=impute_zeros_with_average(df_Oct,3)
```

### Extrapolation

```python
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.linear_model import LinearRegression
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.linear_model import Ridge,Lasso
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import  RBF,WhiteKernel
from sklearn.svm import SVR
from sklearn.ensemble import  RandomForestRegressor,GradientBoostingRegressor
```

```python
data = pd.DataFrame({'Year': np.arange(1986,2020), 'Groundwater_Level': df_punjab.iloc[0,1:].astype(float)})
data.set_index('Year', inplace=True)
```

```python
lr=GradientBoostingRegressor()
lr.fit(np.arange(1986,2020).reshape(-1,1),data['Groundwater_Level'])
```

```python
lr.predict(np.arange(1986,2020).reshape(-1,1))
```

```python
lr.predict(np.arange(2020,2036).reshape(-1,1))
```

```python
future_years = np.arange(2020, 2036)
forecasted_values=lr.predict(np.arange(2020,2036).reshape(-1,1))
extrapolated_data = pd.DataFrame({'Year': future_years, 'Groundwater_Level': forecasted_values})
extrapolated_data.set_index('Year', inplace=True)
```

```python
all_data = pd.concat([data, extrapolated_data])

# Plot the original data and extrapolation
plt.figure(figsize=(10, 6))
plt.plot(all_data.index, all_data['Groundwater_Level'], label='Actual Data + Forecast')
plt.scatter(data.index, data['Groundwater_Level'], color='red', label='Actual Data')
plt.scatter(extrapolated_data.index, extrapolated_data['Groundwater_Level'], color='green', label='Forecast')
plt.xlabel('Year')
plt.ylabel('Groundwater Level')
plt.title('Groundwater Level Extrapolation')
plt.legend()
plt.grid(True)

plt.show()
```

SARIMA

```python
order = (1, 1, 1)  # The order of SARIMA (p, d, q)
seasonal_order = (1, 0, 1, 12)  # Seasonal order (P, D, Q, S)
model = SARIMAX(data['Groundwater_Level'], order=order, seasonal_order=seasonal_order)
results = model.fit()
```

```python
future_years = np.arange(2020, 2035)
forecast = results.get_forecast(steps=len(future_years))
forecasted_values = forecast.predicted_mean
```

```python
extrapolated_data = pd.DataFrame({'Year': future_years, 'Groundwater_Level': forecasted_values})
extrapolated_data.set_index('Year', inplace=True)
```
